```python
data = np.ones(4)
data[2] = 5
data
```

```python
row_indices = np.arange(4)
```

```python
col_indices = np.arange(4)
```

```python
coo = sparse.coo_matrix((data, (row_indices, col_indices)))
coo
```

```python
print(coo)
```

```python
# coo_matrix (coordinated matrix) is used to construct a sparse matrix 
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
y = np.sin(x)
plt.plot(x, y, marker='x')
```

```python
# np.linspace(-10, 10, 100) returns an array of 100 equally spaced elements from -10 to 100 
# plt.plot() will plot all the points taking x and y as arguments (we can also specify marker) and it will join 
# all the points with straight lines on the graph
```
