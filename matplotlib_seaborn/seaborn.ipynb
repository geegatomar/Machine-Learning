```python
import seaborn as sns

df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.shape
```

```python
df.corr()
```

```python
# correlation between each columns with every other. Correlation is calculated for all int/float values
```

```python
# correlation values lie between 0 and 1
```

```python
# sns.heatmap() will give the visualization matrix, which takes the matrix of correlations as input
```

```python
sns.heatmap(df.corr())
```

```python
# This correlation is important because when doing feature extractiion, if two features have very high correlation,
# if correlation is say more that 0.95, then one of them is advised to be dropped.
```

```python
# To analyse 2 features, use JOINTPLOT

sns.jointplot(x="tip", y="total_bill", data=df)
```

```python
# The histograms are present in light blue on both sides for each of those features respectively
```

```python
sns.jointplot(x="tip", y="total_bill", data=df, kind="reg")
```

```python
# To analyse more than 2 features, we use a PAIRPLOT
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue="sex")

# we can provide hue like this
```

```python
sns.pairplot(df, hue="time")
```

```python
# Distplot helps you with the histogram
```

```python
sns.distplot(df["tip"])
```

```python
sns.distplot(df["tip"], kde=False, bins=10)

# on making kde false, it shows the count in y axis. With kde true it was showing y axis in densities
# we can also specify bins
```
