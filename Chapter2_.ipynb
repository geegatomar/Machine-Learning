## Gradient Boost

```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import GradientBoostingClassifier

cancer = load_breast_cancer()
X, y = cancer.data, cancer.target
X_train, X_test, y_train, y_test = train_test_split(X, y)

model = GradientBoostingClassifier().fit(X_train, y_train)
print(model.score(X_test, y_test), model.score(X_train, y_train))
```

```python
model = GradientBoostingClassifier(learning_rate=0.01, n_estimators=250).fit(X_train, y_train)
print(model.score(X_test, y_test), model.score(X_train, y_train))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

n = cancer.data.shape[1]
plt.barh(np.arange(n), model.feature_importances_)
plt.yticks(np.arange(n) , cancer.feature_names)
```

```python
model = GradientBoostingClassifier().fit(X_train, y_train)
print(model.score(X_test, y_test), model.score(X_train, y_train))

plt.barh(np.arange(n), model.feature_importances_)
plt.yticks(np.arange(n) , cancer.feature_names)
```

```python
model = GradientBoostingClassifier(max_depth=1).fit(X_train, y_train)
print(model.score(X_test, y_test), model.score(X_train, y_train))

plt.barh(np.arange(n), model.feature_importances_)
plt.yticks(np.arange(n) , cancer.feature_names)
```
