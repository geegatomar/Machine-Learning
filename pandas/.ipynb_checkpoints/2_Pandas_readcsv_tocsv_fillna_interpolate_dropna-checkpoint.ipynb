## Read from csv file

```python
# pd.read_csv() to read from csv file
```

```python
import pandas as pd

df = pd.read_csv('stock_data.csv', sep="\t", na_values=['not available', 'n.a.'])
df
```

```python
# na_values can be given a list which is replaces with NaN values
```

```python
df = pd.read_csv('stock_data.csv', sep='\t', na_values={
    'eps':['not available', 'n.a.'],
    'revenue':[-1, 'not available', 'n.a.'],
    'price': ['not available', 'n.a.'],
    'people':['not available', 'n.a.']
})
```

```python
df
```

```python
# we can also provide a dictionary as above
```

## Handling missing data in Pandas

### fillna : To fill missing values using diff ways
### interpolate: to make a guess on missing values using interpolation
### dropna: to drop rows will missing values

```python
df = pd.read_csv('weather_data.csv', sep='\t')
```

```python
df
```

```python
df['day']
```

```python
type(df['day'][0])
```

```python
# Its type is string, we need to convert it to date
```

```python
df['day'] = pd.to_datetime(df['day'])
```

```python
df
```

```python
type(df.day[0])
```

```python
# Method 2 to do the same thing is:
```

```python
dff = pd.read_csv('weather_data.csv', sep='\t', parse_dates=['day'])
```

```python
dff
```

```python
df
```

```python
df.fillna(0)

# To replace NaN values with this value (here - 0)
```

```python
df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': 'no event'
})
```

```python
# carry forward values for na values

df.fillna(method="ffill")       # forward fill
```

```python
df.fillna(method="bfill")
```

```python
df
```

```python
df.interpolate()
# default is linear interpolation
```

```python
df.interpolate(method="time")

#this gives error to work only when the index columns is having "date" type values
```

```python
df.set_index('day', inplace=True)
df
```

```python
df.interpolate(method="time", inplace=True)
df
```

```python
# Powerful feature to make guesses for missing values
```

```python
df.dropna()

# drops all rows with NaN values
```

```python
df.dropna(thresh=1)
# allows 1 NaN vaue
```

```python
import numpy as np

np.nan
```

```python
dt = pd.date_range("2017-01-01", "2017-01-11")
idx = pd.DatetimeIndex(dt)
new_df = df.reindex(idx)
new_df
```

```python
new_df.interpolate(method="time")
```

```python
new_df.fillna(method="ffill")
```
