```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = [1, 4, 7, 9, 10]
y = [4, 5, 10, 1, 2]
plt.plot(x, y)
```

```python
plt.xlabel('Day')
plt.ylabel('Temp')
plt.title('Weather data')

plt.plot(x, y, marker="o", linestyle="dashed", linewidth=3)
```

```python
plt.plot(x, y, '--r*')
```

```python
plt.plot(x, y, '*g-.')
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[4,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="avg")

plt.legend()
```

```python
plt.bar(x, y)
```

```python
plt.hist(x, bins = [0, 4, 9, 11, 13], rwidth=0.9)
```

```python
plt.pie(x, radius=2, autopct="%0.2f%%")
```
